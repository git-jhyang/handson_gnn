- regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

num_data = 5

np.random.seed(123)
x = np.random.rand(100) * 10
y = x + np.random.normal(0, 1, 100)
model = LinearRegression()
model.fit(x[:num_data].reshape(-1,1), y[:num_data].reshape(-1,1))

px = [[0], [10]]
py = model.predict(px)

f, ax = plt.subplots(1,1,figsize=(6,4))
ax.scatter(x,y, label='All data')
ax.scatter(x[:num_data], y[:num_data], label='Trained data')
ax.plot([0,10],[0,10], color='k', label='Truth: y=x')
a, b = model.coef_, model.intercept_
ax.plot(px,py, color='r', label=f'Predict: y={a[0,0]:.3f}x{b[0]:+.3f}')
ax.legend(loc='upper left')
```
